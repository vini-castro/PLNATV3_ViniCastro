#### **UTILIZAÇÃO BÁSICA**

Nesse código, são feitas várias requisições para a API, que retorna os dados de vídeos populares aletórios. Por essa requisição, não é possível selecionar quantos vídeos serão retornados, então são feitas várias requisições, e elas são armazenadas em um vetor. Desse modo, é consideravelmente mais lento, mas funciona para pegar vídeos o suficiente, para se ter uma base com uma boa qualidade. Se um erro for detectado na conexão, o programa não vai ser interrompido, ele apenas não concatenará os dados à lista.

```
palavras_ignoradas = ['que', 'de','do','da', 'a', 'o', 'e', 'para', 'um', 'uma', 'com', 'em', 'no', 'na','é','não','se']

id_list = []
for _ in range(20):
  conn.request("GET", "/feed?region=BR&device_id=7523368224928586621", headers=headers)

  res = conn.getresponse()
  data = res.read()

  json_data = data.decode("utf-8")
  try:
    data_dict = json.loads(json_data)
    num_aweme = len(data_dict['aweme_list'])
    for i in range(num_aweme):
      aweme_id = data_dict['aweme_list'][i]['desc']
      id_list.append(aweme_id)
  except json.JSONDecodeError as e:
    print(f"Erro ao decodificar JSON: {e}")
    continue

all_descriptions = ' '.join(id_list)

words = re.findall(r'\b\w+\b', all_descriptions.lower())

words_filtradas = [word for word in words if word not in palavras_ignoradas]

word_counts = Counter(words_filtradas)

sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
```

#### **EXEMPLOS PRÁTICOS**

Aqui pode ser visto um exemplo prático, em que é criado um word cloud para verificar quais são as palavras que mais aparecem nas explicações e descrições dos vídeos.

```
palavras_nuvem = {word: count for word, count in sorted_words if count >= 2}

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(palavras_nuvem)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
